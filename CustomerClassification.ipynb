```python
# Importing the Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('customers.csv')
df
```

```python
# Finding for null values
df.isnull().sum()
```

```python
X = df.iloc[:,[3,4]].values
```

```python
X
```

```python
# finding wcss value for different number of clusters

wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=2)
  kmeans.fit(X)

  wcss.append(kmeans.inertia_)
```

```python
# plot an elbow graph

sns.set()
plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# We got an elbow at five so we select the value of number of cluster = five
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=2)
# return a label for each data point based on their cluster
Y = kmeans.fit_predict(X)
```

```python
# plotting all the clusters and their Centroids

plt.figure(figsize=(4,4))
plt.scatter(X[Y==0,0], X[Y==0,1], s=50, c='green', label='Cluster 1')
plt.scatter(X[Y==1,0], X[Y==1,1], s=50, c='red', label='Cluster 2')
plt.scatter(X[Y==2,0], X[Y==2,1], s=50, c='black', label='Cluster 3')
plt.scatter(X[Y==3,0], X[Y==3,1], s=50, c='magenta', label='Cluster 4')
plt.scatter(X[Y==4,0], X[Y==4,1], s=50, c='blue', label='Cluster 5')

# plot the centroids
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='brown', label='Centroids')

plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
X = (15,81)
X = np.asarray(X)
X = X.reshape(1,-1)
pred = kmeans.predict(X)[0]
if pred == 0:
  print("It belongs to cluster 1")
elif pred == 1:
  print("It belongs to cluster 2")
elif pred == 2:
  print("It belongs to cluster 3")
elif pred == 3:
  print("It belongs to cluster 4")
if pred == 4:
  print("It belongs to cluster 5")
```
